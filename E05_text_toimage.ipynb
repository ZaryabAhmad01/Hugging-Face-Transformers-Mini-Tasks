```python
from diffusers import StableDiffusionPipeline
import torch

# ==========================
# 1. LOAD THE MODEL PIPELINE
# ==========================
model_id = "CompVis/stable-diffusion-v1-4"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16      # Use half precision to save GPU memory
)

# Move the pipeline to GPU if available, otherwise CPU
device = "cuda" if torch.cuda.is_available() else "cpu"
pipe = pipe.to(device)
```

```python
# ==========================
# 2. DEFINE YOUR TEXT PROMPT
# ==========================
# prompt = "Flying cars soar over a futuristic cityscape at sunset."

prompt = "generate a image as on saudia a camel stnad in the sand and with arabic people."

```

```python
# ==========================
# 3. GENERATE THE IMAGE
# ==========================
# 'autocast' automatically manages mixed precision (faster on modern GPUs)
with torch.autocast(device):
    image = pipe(prompt).images[0]   # pipe returns a list; take the first image

```

```python
# ==========================
# 4. SAVE THE IMAGE
# ==========================
image.save("generated_image.png")
print("Image saved as generated_image.png")
```
