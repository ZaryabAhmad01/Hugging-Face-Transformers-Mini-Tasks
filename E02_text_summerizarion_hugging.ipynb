```python
from transformers import AutoModelForSeq2SeqLM , AutoTokenizer

```

```python
model_name = "facebook/bart-large-cnn"  # note: lowercase "large" is standard
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
# Input text to summarize (fixed: previously empty string)
text = """
Virat Kohli is an Indian international cricketer known for his exceptional batting skills and he has played for the Indian national team in all formats and is a former captain. Kohli is referred to as "King Kohli" and "Chase Master" due to his remarkable ability to chase down targets.
"""

# Tokenize the input text with a task prefix
inputs = tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=512, truncation=True)

# Generate the summary
summary_ids = model.generate(inputs, max_length=50, min_length=25, length_penalty=2.0, num_beams=1)
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

print(summary)
```
